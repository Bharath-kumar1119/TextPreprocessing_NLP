## Stemming
- Process of reducing the word to its word stem.
### Classification problem
#### Looking for the positive reviews and negative reviews.
- (eat, eating, eaten)---> eat 
- (go, goes, going)----> go

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

## PorterStemmer

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

### RegexpStemmer class
NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|E$|ablE$|ul$', min = 4)
```

```python
print(reg_stemmer.stem('understanding'))
print(reg_stemmer.stem('cats'))
print(reg_stemmer.stem('adorablE'))
print(reg_stemmer.stem('understandable'))
print(reg_stemmer.stem('ul'))
print(reg_stemmer.stem('ululul'))
```

### Snowball Stemmer
 It is a stemming algorithm which is also known as the Porter2 stemming algorithm as it is a better version of the Porter Stemmer since some issues of it were fixed in this stemmer.

```python
from nltk.stem import SnowballStemmer
snow = SnowballStemmer('english')
snow.stem('cats')
```

```python
for word in words:
    print(word+"--->"+snow.stem(word))
    print(word+"--->"+stemming.stem(word))
    print("-------------------")
```

```python
stemming.stem('fairly'), stemming.stem('sportingly')
```

```python
snow.stem('fairly'), snow.stem('sportingly')
```

```python
snow.stem('goes')
```

```python
stemming.stem('comfortable')
```

```python
snow.stem('comfortable')
```

```python
for word in words:
    print(word+"--->"+snow.stem(word))
    print(word+"--->"+snow1.stem(word))
    print("-------------------")
```
