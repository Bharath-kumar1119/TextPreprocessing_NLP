```python
corpus = """Hello Welcome, to Bharath Kumar's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
### Tokenization
## paragraphs   ---> Sentence 

from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt_tab')
```

```python
documents = sent_tokenize(corpus)
```

```python
print(type(documents))
```

```python
for sentence in documents:
    print(sentence)
```

```python
### Tokenization
## Paragraph ---> Words
## Sentences --> Words
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```

```python
para = """
The Indian Space Research Organisation (ISRO) announced a new mission on Tuesday.
The project aims to study climate patterns using advanced satellite imaging.
Experts believe this mission could significantly improve weather forecasting in the country.
"""
```

```python
print(para)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize, wordpunct_tokenize, TreebankWordTokenizer
documents = sent_tokenize(para)
for sentence in documents:
    print(sentence)

for sentence in documents:
    print(word_tokenize(sentence))

print(wordpunct_tokenize(para))
print(tokenizer.tokenize(para))

```
